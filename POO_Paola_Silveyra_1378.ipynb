#### Conceptos de POO (Programacion Orientada a Objetos) en Python

```python
# El constructor: se ejecuta al crear el objeto
class Animal:
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    def ladrar(self):
        print(f"{self.nombre} (raza {self.raza}) dice: ¡Guau, guau!")

    def correr(self):
        print(f"{self.nombre} está corriendo a toda velocidad.")

# Creación de los objetos
pitbull = Animal("Roco", "Pitbull")
chihuahua = Animal("Chispita", "Chihuahua")

# Uso de las funciones
pitbull.ladrar()
pitbull.correr()

print("-" * 20)

chihuahua.ladrar()
chihuahua.correr()
print("Programa realizado por Paola Silveyra NL = 50")
```

```python
class Animal:
    # Constructor: inicializa los atributos al crear el objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para ladrar
    def ladrar(self):
        print(f"{self.nombre} (de raza {self.raza}) dice: ¡Guau, guau!")

    # Función para correr
    def correr(self):
        print(f"{self.nombre} está corriendo rápidamente. 🐵​")
    def morder(self):
        print(f"{self.nombre} está mordiendo a la vecina 🙎‍♀️​")

# Creación de los objetos (instancias de la clase Animal)
pitbull = Animal("Rocco", "Pitbull")
chihuahua = Animal("Taco", "Chihuahua")
# creamos al objeto perro calejero con datos
perrocallejero = Animal("solovino", "chafa")
# Uso de los objetos
pitbull.ladrar()
pitbull.correr()

print("-" * 20)

chihuahua.ladrar()
chihuahua.correr()

print("-" * 20)

perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.morder()
print("programa realizado por Paola Silveyra NL = 50")
```
